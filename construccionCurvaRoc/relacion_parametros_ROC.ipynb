```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_parametros_ROC = pd.read_csv('AUC_vs_activos.txt', sep = ',')
print(df_parametros_ROC)
```

```python
plt.figure(figsize=(15,7))
plt.scatter(x=df_parametros_ROC['NUMERO'], y=df_parametros_ROC['AUC'])
plt.xlabel('NUMERO DE ACTIVOS', fontsize=20)
plt.ylabel('AUC', fontsize=20)
```

```python
plt.figure(figsize=(15,7))
plt.plot(df_parametros_ROC['NUMERO'], df_parametros_ROC['ENRINCHMENT_FACTOR'])
plt.xlabel('NUMERO DE ACTIVOS', fontsize=20)
plt.ylabel('ENRINCHMENT_FACTOR', fontsize=18)
```
